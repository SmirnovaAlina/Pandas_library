```python
# Dependencie
import pandas as pd 
import numpy as np 
```

```python
# Read and display the CSV with Pandas
file = "Resources/purchase_data.csv"
df = pd.read_csv(file)
df.head(10)
```

```python
# Identify incomplete rows
df.count()
```

```python
# Total Number of Players
total_ofPlayers = len(df["SN"].unique())
len_total = len(df["SN"])
print(len_total)
total_ofPlayers  
```

```python
# Number of Unique Items
number_ofItems = len(df["Item ID"].unique())
number_ofItems
```

```python
# Average Purchase Price
mean_ofPrice = round(df["Price"].mean(), 2)
mean_ofPrice
```

```python
# Number of Purchases
number_ofPurchases = df["Purchase ID"].count()
number_ofPurchases
```

```python
# Total Revenue
Total_revenue = df["Price"].sum()
Total_revenue 
```

```python
# Purchasing Analysis
purchasing_sum = pd.DataFrame({"Number of Unique Items": number_ofItems, "Average Purchase Price": mean_ofPrice, "Number of Purchases": number_ofPurchases,
                              "Total Revenue": [Total_revenue] })
# purchasing_sum["Average Purchase Price"] = purchasing_sum["Average Purchase Price"].map("${:.2f}".format)
# purchasing_sum["Total Revenue"] = purchasing_sum["Total Revenue"].map("${:.2f}".format)

purchasing_sum
```

```python
# Gender Demographics

Gender_demographics = round(df['Gender'].value_counts(normalize=True) * 100 , 2)
Gender_demographics = pd.DataFrame(Gender_demographics)
Gender_demographics['Gender'] = Gender_demographics['Gender'].map("%{:.2f}".format)
Gender_demographics
```

```python
# Female analysis 
female_df = df.loc[df["Gender"] == "Female", :]
f_purchaseCount = female_df['Purchase ID'].count()
f_priceMean = round(female_df['Price'].mean(), 2)
f_purchaseValue = female_df['Price'].sum()
unique_person_w = len(female_df['SN'].unique())
average_perWomen = round(f_purchaseValue/unique_person_w, 2)

print(f_purchaseCount)
print(f_priceMean)
print(f_purchaseValue)
print(unique_person_w)
average_perWomen
```

```python
# Male analysis
male_df = df.loc[df["Gender"] == "Male", :]
m_purchaseCount = male_df['Purchase ID'].count()
m_priceMean = round(male_df['Price'].mean(), 2)
m_purchaseValue = male_df['Price'].sum()
unique_person_m = len(male_df['SN'].unique())
average_perMen = round(m_purchaseValue/unique_person_m, 2)

print(m_purchaseCount)
print(m_priceMean)
print(m_purchaseValue)
print(unique_person_m)
print(average_perMen)
```

```python
others_df = df.loc[df["Gender"] == "Other / Non-Disclosed", :]
o_purchaseCount = others_df['Purchase ID'].count()
o_priceMean = round(others_df['Price'].mean(), 2)
o_purchaseValue = others_df['Price'].sum()
unique_person_o = len(others_df['SN'].unique())
average_perOthers = round(o_purchaseValue/unique_person_o, 2)
print(o_purchaseCount)
print(o_priceMean)
print(o_purchaseValue)
print(unique_person_o)
print(average_perOthers)
```

```python
Purchases_Analysis = pd.DataFrame({"Purchase Count" : [f_purchaseCount, m_purchaseCount, o_purchaseCount ], "Average Purchase Price": [f_priceMean, m_priceMean, o_priceMean ], "Total Purchase Value": [f_purchaseValue, m_purchaseValue, o_purchaseValue],
                                  "Avg Total Purchase per Person": [average_perWomen, average_perMen, average_perOthers],
                                  "Gender" : ["Female", "Male", "Other / Non-Disclosed"]})
# Purchases_Analysis["Average Purchase Price"] = Purchases_Analysis["Average Purchase Price"].map("${:.2f}".format)
# Purchases_Analysis["Total Purchase Value"] = Purchases_Analysis["Total Purchase Value"].map("${:.2f}".format)
# Purchases_Analysis["Avg Total Purchase per Person"] = Purchases_Analysis["Avg Total Purchase per Person"].map("${:.2f}".format)
Purchases_Analysis.set_index("Gender")

# Purchases_Analysis
```

```python
# Age Demographics
# Find max and min value to understand a range 
age_max = df["Age"].max()
age_min = df["Age"].min()
print(age_max)
print(age_min)
# Create bins 
bins = [4, 9, 14, 19, 24, 29,
        34, 39, 45]
# Create labels for these bins
bins_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40+"]
```

```python
# Slice the data and place it into bins
pd.cut(df["Age"], bins, labels=bins_labels).head(10)
df["View Age Group"] = pd.cut(df["Age"], bins, labels=bins_labels)
df.head(10)
```

```python
# Create a GroupBy object based upon "View Age Group"
df_group = df.groupby("View Age Group")
# Find how many unique rows fall into each bin
count_group_unique = df_group["SN"].nunique()
print(count_group_unique)
# Find percantage to the Total of Players
percentage_of_p = round((count_group_unique/total_ofPlayers)*100, 2)
print(percentage_of_p)
```

```python
# Data Frame with Age Demographics analysis

summary_age_demographics = pd.concat([count_group_unique, percentage_of_p,], axis=1)
summary_age_demographics.columns = ["Total Count","Percentage of Players"]
summary_age_demographics
```

```python
# Find how many (not unique) rows fall into each bin
purchase_count = df_group["View Age Group"].count()
print(purchase_count)
# Find average price by the group
average_total_price = round(df_group["Price"].mean(),2)
print(average_total_price)
# Find Total Purchase Value
Total_Pvalue_bygroup = round(df_group["Price"].sum(),2)
print(Total_Pvalue_bygroup)
# Average Purchase Total per Person by Age Group
Aver_Total_per_person = round(Total_Pvalue_bygroup/count_group_unique,2)
print(Aver_Total_per_person)
```

```python
#  Purchasing Analysis (Age)
summary_age_purchasing_analysis = pd.concat([purchase_count, average_total_price, Total_Pvalue_bygroup, Aver_Total_per_person], axis=1)

summary_age_purchasing_analysis.columns = ["Purchase Count","Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]                                                                             
summary_age_purchasing_analysis
```

```python
# Top Spenders
groupby_spenders = df.groupby("SN")
sum_by_spenders = groupby_spenders["Price"].sum()
purchase_count = groupby_spenders["Item Name"].count()
average_pp = round((groupby_spenders["Price"].mean()),2)

```

```python
#  Top Spenders Analysis

top_spenders = pd.concat([purchase_count, average_pp, sum_by_spenders], axis=1)
top_spenders.columns = ["Purchase Count","Average Purchase Price", "Total Purchase Value" ]
# top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${:.2f}".format)
# top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${:.2f}".format)
top_spenders.sort_values("Total Purchase Value", ascending=False).head(5)

```

```python
# Identify the 5 most popular items by purchase count
groupby_item = df.groupby(["Item ID", "Item Name","Price"])
sum_by_items = groupby_item["Price"].sum()
item_purchase_count = groupby_item["Item Name"].count()
```

```python
#  Most Popular Items Analysis

top_items = pd.concat([item_purchase_count, sum_by_items], axis=1)
top_items.columns = ["Purchase Count", "Total Purchase Value" ]
# top_items["Total Purchase Value"] = top_items["Total Purchase Value"].map("${:.2f}".format)

top_items.sort_values("Purchase Count", ascending=False).head(5)
```

```python
# Most Profitable Items
top_items = pd.concat([item_purchase_count, sum_by_items], axis=1)
top_items.columns = ["Purchase Count", "Total Purchase Value" ]
top_items.sort_values("Total Purchase Value", ascending=False).head(5)
```
